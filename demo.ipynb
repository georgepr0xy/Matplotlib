```python
import matplotlib.pyplot as plt
x= [1,2,3,4]
y= [3,9,27,81]

plt.xlabel("language" , fontsize = 20 , color = "turquoise")
plt.ylabel("popularity" , fontsize = 20, color = "limegreen")
plt.title("Barplot", fontsize =20 , color = "brown")
plt.bar(x,y, color = ["r","g","y","b"], width = 0.4, align ="edge", edgecolor= "black", linewidth = 5.0, linestyle = ":", alpha = 0.7)
plt.show()
```
