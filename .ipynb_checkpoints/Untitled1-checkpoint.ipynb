Bar plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

```python
x=["python","C","c++","java"]
y=[85,70,60,82]
c = ["r","g","y","b"]
plt.xlabel("language" , fontsize = 16)
plt.ylabel("popularity" , fontsize = 16)
plt.title("George barplot")

plt.bar(x,y, color = c,align = "center", width = 0.5)
plt.show()
```
