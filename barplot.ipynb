Bar plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

```python
x=["python","C","c++","java"]
y=[85,70,60,82]
z = [20, 30, 40 ,50]
c = ["r","g","y","b"]
plt.xlabel("language" , fontsize = 16)
plt.ylabel("popularity" , fontsize = 16)
plt.title("George barplot")

plt.bar(x,y, color = "y",align = "center", width = 0.5, label = "popularity")
plt.bar(x,z, color = "r",align = "center", width = 0.5, label = "popularity1")
plt.legend()
plt.show()
```

```python
x=["python","C","c++","java"]
y=[85,70,60,82]
z = [20, 30, 40 ,50]


width = 0.1
p = np.arange(len(x))
p1 =[j+width for j in p]


c = ["r","g","y","b"]
plt.xlabel("language" , fontsize = 16)
plt.ylabel("popularity" , fontsize = 16)
plt.title("George barplot", fontsize = 20)
plt.xticks(p+width/2,x,rotation= 20)

plt.bar(p,y, color = "y",align = "center", width = 0.5, label = "popularity")
plt.bar(p1,z, color = "r",align = "center", width = 0.5, label = "popularity1")
plt.legend()
plt.show()
```

```python
x=["python","C","c++","java"]
y=[85,70,60,82]
z = [20, 30, 40 ,50]


width = 0.1
p = np.arange(len(x))
p1 =[j+width for j in p]


c = ["r","g","y","b"]
plt.xlabel("language" , fontsize = 16)
plt.ylabel("popularity" , fontsize = 16)
plt.title("George barplot", fontsize = 20)
plt.xticks(p+width/2,x,rotation= 20)

plt.barh(x,y, color = "y",  label = "popularity")
plt.barh(x,z, color = "r",  label = "popularity1")
plt.legend()
plt.show()
```
